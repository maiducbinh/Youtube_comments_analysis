# Ph√¢n lo·∫°i n·ªôi dung comment

```python
import openai
import pandas as pd
```

```python
# ƒê·ªçc API key t·ª´ file
with open("apikey.txt", "r") as f:
    openai.api_key = f.readline()
```

```python
# ƒê·ªãnh nghƒ©a h√†m ƒë·ªÉ ph√¢n lo·∫°i c√°c b√¨nh lu·∫≠n
def classify_comment(comment):
    # G·ªçi API ƒë·ªÉ ph√¢n lo·∫°i b√¨nh lu·∫≠n
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=f"Is this comment positive or negative?\nComment: {comment}\nLabel:",
        temperature=0.5,
        max_tokens=1,
        n=1,
        stop=None,
    )

    # L·∫•y k·∫øt qu·∫£ tr·∫£ v·ªÅ t·ª´ API
    label = response.choices[0].text.strip().lower()

    # Tr·∫£ v·ªÅ 1 n·∫øu b√¨nh lu·∫≠n ti√™u c·ª±c, ng∆∞·ª£c l·∫°i tr·∫£ v·ªÅ 0
    return 1 if label == "negative" else 0
```

- model: "text-davinci-003"
- max_tokens = 1: 'negative' or 'positive'

```python
# Th√™m c·ªôt "label" cho dataframe b·∫±ng c√°ch √°p d·ª•ng h√†m classify_comment cho t·ª´ng b√¨nh lu·∫≠n
df["label"] = df["text"].apply(classify_comment)

# In ra danh s√°ch c√°c author c√≥ b√¨nh lu·∫≠n ti√™u c·ª±c
negative_authors = df[df["label"] == 1]["author"].unique()
print("Negative authors:")
print(negative_authors)
#print(openai.api_key)
```
